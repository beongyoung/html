```python
import matplotlib.pyplot as plt
```

# matplotlib
    - 2D 그래프로 시각화가 가능한 라이브러리
    - 데이터를 차트나 plot으로 그려줌
    - 차트의 정리 : 산정도, 바 차트, 파이, 히스토그램, 꺾은선 그래프 등 다양함

### 데이터 시각화
    - 광범위하게 분산된 방대한 양의 자료를 한 눈에 볼 수 있도록 도표나 차트로 정리하는 것

```python
# X축과 Y 축을 지정해서 선 그래프를 그리기
x = [1,2,3,4,5]
y = [1,5,7,8,6]

# plot은 선 그래프를 그려주는 명령어
# plot(x axis, y axis, ls(linestyle) = '')
plt.plot(x,y, ls = '--')

# 그래프 출력
plt.show()
```

```python
# X축과 Y 축을 지정해서 선 그래프를 그리기
x = [1,2,3,4,5]
y = [1,5,7,8,6]

# plot은 선 그래프를 그려주는 명령어
# plot(x axis, y axis, ls(linestyle) = '')
plt.plot(x,y, ls = '--', lw = 5)

# 그래프 출력
plt.show()
```

```python
import numpy as np
import numpy.random as rd
```

```python
x = np.arange(7)
y = [1,5,7,8,6,9,4]
# mec : 마커 포인트 강조 색상
# mfc : 마커 채울 색상
# ms : 마커의 크기
plt.plot(x,y, ls = ':', marker = 'd', c = 'm', ms = 7, mfc = 'b')
plt.show()
```

```python
x = np.arange(7)
y = [1,4,5,8,9,5,3]
# mec : 마커 포인트 강조 색상
# mfc : 마커 채울 색상
# ms : 마커의 크기
plt.plot(x,y, ls = '--', lw = 5, marker = 'o', c = 'b', mew = 5, mec = 'g', ms = 16, mfc = 'r')
plt.show()
```

```python
x = np.arange(7)
y = [1,4,5,8,9,5,3]

plt.plot(x,y, ls = '--', lw = 5, marker = 'o', c = 'b', mew = 5, mec = 'g', ms = 16, mfc = 'r')
plt.xlim(-2, 10)
plt.ylim(-5, 12)
plt.show()
```

### tic 설정
    - 그래프 x축, y축 상의 위치 표시 지정을 틱이라고 함.
    - 틱은 일반적으로 matplotlib에서 자동으로 정해줌.
    - 수동으로도 설정가능

```python
x = np.arange(7)
y = [1,4,5,8,9,5,3]

plt.plot(x,y, ls = '--', lw = 5, marker = 'o', c = 'b', mew = 5, mec = 'g', ms = 16, mfc = 'r')
plt.xticks([0,3,6])
plt.yticks([1,5,9])
plt.show()
```

### Grid
    - 그림 중간에 선을 자동으로 그려주는 것

```python
x = np.arange(7)
y = [1,4,5,8,9,5,3]

plt.plot(x,y, ls = '--', lw = 5, marker = 'o', c = 'b', mew = 5, mec = 'g', ms = 16, mfc = 'r')
plt.grid()
plt.show()
```

```python
x = np.arange(7)
y = [1,4,5,8,9,5,3]

x1 = np.arange(10)
y1 = np.arange(10)

plt.plot(x,y, ls = '--', lw = 5, marker = 'o', c = 'b', mew = 5, mec = 'g', ms = 16, mfc = 'r')
plt.plot(x1,y1, ls = ':')

plt.show()
```

### Pie 차트

```python
label = ['개구리', '돼지', '개', '통나무']
sizes = [15, 30, 45, 10]
colors = ['yellowgreen', 'gold', 'lightskyblue', 'lightcoral']
explode = (0, 0.1, 0, 0)
plt.title("Pie Chart")

plt.pie(sizes, explode = explode, labels = label, colors = colors, 
        autopct = '%1.1f%%', shadow = True, startangle = 90)
plt.axis('equal')
plt.show()
```

```python
rd.seed(1)
x = rd.normal(0, 1, 100)
y = rd.normal(0, 1, 100)
plt.title("Scatter Plot")

plt.scatter(x, y)
plt.show()
```
